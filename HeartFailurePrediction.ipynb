#Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#Importing Dataset

```python
dataset = pd.read_csv('heart.csv')
X = dataset.iloc[:,:-1].values  #here in iloc fn first parameter : means taking all rows and 2nd para :-1 means taking all columns except the last column
Y = dataset.iloc[:,-1].values   #here in iloc fn first parameter : means taking all rows and 2nd para -1 means only taking the last column
```

```python
print(X)
```

```python
print(Y)
```

#Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder      #we are using label encoding here to encode the features which are in the form of strings
le1 = LabelEncoder()
le2 = LabelEncoder()
le6 = LabelEncoder()
le8 = LabelEncoder()
le10 = LabelEncoder()

X[:,1] = le1.fit_transform(X[:,1])
X[:,2] = le2.fit_transform(X[:,2])
X[:,6] = le6.fit_transform(X[:,6])
X[:,8] = le8.fit_transform(X[:,8])
X[:,10] = le10.fit_transform(X[:,10])
```

```python
print(X)
```

#Splitting the dataset into Training set and Testing set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(Y_train)
```

```python
print(Y_test)
```

###In the above 4 print statements you can see in the output that the dataset is divided into 80% train set and 20% test set

#Feature Scaling
### It is done to make the values lie in a particular range so that model becomes faster and graph is plotted easily

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
# we are going to perform feature scaling only on xtrain and xtest as in y there are only 0s and 1s so no need to scale y
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
print(X_train)  #scaled values
```

```python
print(X_test) #scaled values
```

#Training the dataset

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
model_logistic = LogisticRegression()
model_logistic.fit(X_train,Y_train)
```

```python
#Support Vector Machine (SVM)
from sklearn.svm import SVC
model_svm = SVC()
model_svm.fit(X_train,Y_train)
```

```python
#K-Nearest Neighbours (KNN)
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier(n_neighbors=8)
model_knn.fit(X_train,Y_train)
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier
model_tree = DecisionTreeClassifier()
model_tree.fit(X_train,Y_train)
```

```python
# Random Forest Algorithm
from sklearn.ensemble import RandomForestClassifier
model_random = RandomForestClassifier(n_estimators=200)
model_random.fit(X_train,Y_train)
```

```python
# XGBoost
from xgboost import XGBClassifier
model_xgb = XGBClassifier()
model_xgb.fit(X_train,Y_train)
```

```python
y_pred_logistic = model_logistic.predict(X_test)
y_pred_svm = model_svm.predict(X_test)
y_pred_knn = model_knn.predict(X_test)
y_pred_tree = model_tree.predict(X_test)
y_pred_random = model_random.predict(X_test)
y_pred_xgb = model_xgb.predict(X_test)
```

#Testing accuracy on different ML algorithms

```python
from sklearn.metrics import accuracy_score
Logistic_Accuracy = accuracy_score(Y_test,y_pred_logistic)
SVM_Accuracy = accuracy_score(Y_test,y_pred_svm)
KNN_Accuracy = accuracy_score(Y_test,y_pred_knn)
Tree_Accuracy = accuracy_score(Y_test,y_pred_tree)
Random_Accuracy = accuracy_score(Y_test,y_pred_random)
XGB_Accuracy = accuracy_score(Y_test,y_pred_xgb)
```

```python
print(Logistic_Accuracy)
print(SVM_Accuracy)
print(KNN_Accuracy)
print(Tree_Accuracy)
print(Random_Accuracy)
print(XGB_Accuracy)
```

# Plotting Accuracy of different ML algorithms

```python
plt.figure(figsize=(12,7))
plt.bar("Logistic Regression",Logistic_Accuracy,width=0.5)
plt.bar("Kneighbours",KNN_Accuracy,width=0.5)
plt.bar("SVM",SVM_Accuracy,width=0.5)
plt.bar("Decision Tree",Tree_Accuracy,width=0.5)
plt.bar("Random Forest",Random_Accuracy,width=0.5)
plt.bar("XGBoost",XGB_Accuracy,width=0.5)
plt.xlabel("Machine Learning Algorithm")
plt.ylabel("Accuracy")
```
